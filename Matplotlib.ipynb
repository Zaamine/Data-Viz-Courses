### Tracer une courbe 

#### Une courbe est par définition un ensemble de points de coordonnées (x, y) reliés par une ligne brisée. Plus le nombre de points est important, plus la courbe aura un aspect lisse.

La méthode plot permet de tracer des courbes qui relient des points dont les abscisses et ordonnées sont fournies dans des listes ou tableaux.
Pour tracer un graphe avec les valeurs 'x' en abscisses et les valeurs 'y' en ordonnées on écrit: plt.plot(x,y)

```python
x = [0,2,3,6]
y= [1,4,4,8]
plt.plot(x); #afficher en ordonées ( bleu), donc les ordonnées ont automatiquement pris [0,1,2,3]
plt.plot(x,y); 
plt.title('Le titre de ma courbe');
plt.xlabel("Abscisses");
plt.ylabel("Ordonnées");
```

Il est possible de délimiter les domaines des abscisses et des ordonnées en utilisant la méthode axis.
Elle s'utilise ainsi : axis([xmin, xmax, ymin, ymax]).
Pour fixer indépendamment le domaines des abscisses ou des ordonnées on peut utiliser respectivement xlim([xmin, xmax]) ou ylim([ymin, ymax]).

```python
x = [0,2,3,6]
y= [1,4,4,8]
plt.plot(x); #afficher en ordonées ( bleu), donc les ordonnées ont automatiquement pris [0,1,2,3]
plt.plot(x,y); 
plt.axis([1,3,1,5])
plt.title('Le titre de ma courbe');
plt.xlabel("Abscisses");
plt.ylabel("Ordonnées");
```

```python
plt.plot([0,0,1,1,0,0.5,1],[1,0,0,1,1,2,1])
plt.axis([-1,2,-1,2])
#plt.axis([-5,5,-5,5])
```

La fonction plot fonctionne également avec les Series de Pandas. Lorsque l'on travaille avec un DataFrame, on peut donc facilement en extraire les colonnes pour créer des graphiques. Par exemple, plt.plot(mon_df.col1, mon_df.col2).

Pour superposer plusieurs courbes sur le même graphique, on peut cumuler les instructions plot les unes après les autres, comme ceci :

plt.plot(x,y)
plt.plot(u,v)
plt.plot(w,z)

ou simplement utiliser la syntaxe suivante:

plt.plot(x,y,u,v,w,z).

```python
t = np.arange(0,5,0.2)
plt.plot(t,t,t,t**2,t,t**3)
plt.ylim(0,5)
```

```python
t = np.arange(0,5,0.2)
plt.plot(t,t,t,t**2,t,t**3)
plt.ylim(0,5)
```

Pour chaque paire (x,y) d'arguments, il y a un troisième argument optionnel qui est une chaîne de caractères qui indique le style et la couleur de la ligne de la courbe. Les styles de courbes possibles sont:

'-' : ligne continue
'--' : tirets
':' : ligne en pointillé
'-.' : tirets points
L'ensemble des couleurs possibles est:

'b' (bleu)
'g' (vert) 
'r' (rouge) 
'c' (cyan) 
'm' (magenta) 
'y' (jaune) 
'k' (noir) 
'w' (blanc)`
Par exemple, le format par défaut est ‘b-‘, qui est une ligne continue bleue.

```python
t = np.arange(0,5,0.2)
plt.plot(t,t,'r-.',t,t**2,'g:',t,t**3,'b--')
plt.ylim(0,5)
```

```python
t = np.arange(0,5,0.2)
plt.plot(t,t,'r-.',label = ' linéiaire', linewidth=0.5)
plt.plot(t,t**2,'g:',label = 'carré', linewidth=2)
plt.plot(t,t**3,'b--',label = 'cubique', linewidth=3)
plt.ylim(0,5)
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.legend();

```


## Graphiques en barres avec Barplots




La fonction plt.bar permet de tracer des diagrammes en barres, verticaux avec une seule ou plusieurs séries de valeurs.

Pour afficher un bar plot il suffit d'entrer dans la fonction en premier argument les positions de l'axe des abscisses sur lesquelles les barres seront centrées, et en second argument les hauteurs des barres.

Les paramètres supplémentaires les plus importants sont: la couleur de l'intérieur des barres (color) et la largeur relative de chaque barre (width par défaut 0.8, c'est à dire 20% d'espace vide entre chaque barre).

Comme toutes les fonctions de pyplot il existe beaucoup d'autres paramètres comme :

edgecolor : la couleur des encadrements.
linewidth : l'épaisseur des traits.
yerr : les valeurs des barres d'erreur.
ecolor : la couleur des barres d'erreur.
align: l'axe des abscisses correspond au centre de la barre , ou au bord gauche.
orientation : 'horizontal' pour afficher un barplot horizontal.
hatch = '/' : les hachures. Valeurs possibles : '/', '\', '|', '-', '+', 'x', 'o', 'O', '.', '*'.

```python
plt.bar(range(7),[3,4,5,6,5,4,3],width=0.5,color='b')
```

```python
plt.bar(range(len(df.iphone_sales)),df.iphone_sales,width=1);
```

```python
plt.bar(range(len(df.xiaomi_sales)),df.xiaomi_sales,color='b',label="Vente d'iphones",width=1);
plt.bar(range(len(df.samsung_sales)),df.samsung_sales,color='r',label="Vente de Samsung",width=1);
plt.legend()
plt.xlabel('Semaines')
plt.ylabel('Ventes')
plt.title('Comparaison ventes entre Samsung et Iphone');
```

```python
plt.bar(range(len(df.xiaomi_sales)),df.xiaomi_sales,color='b',label="Vente de Xiaomi",width=1);
plt.bar(range(len(df.samsung_sales)),df.samsung_sales,color='r',label="Vente de Samsung",width=1);
plt.legend()
plt.xlabel('Semaines')
plt.ylabel('Ventes')
plt.ylim(0,1500)
plt.title('Comparaison ventes entre Samsung et Iphone');
```

```python
plt.bar(range(len(df.xiaomi_sales)),df.xiaomi_sales,color='b',label="Vente de Xiaomi",width=1);
plt.bar(range(len(df.samsung_sales)),df.samsung_sales,color='r',bottom=df.xiaomi_sales,label="Vente de Samsung",width=1);
plt.legend()
plt.xlabel('Semaines')
plt.ylabel('Ventes')
plt.ylim(0,1500)
plt.title('Comparaison ventes entre Samsung et Iphone');
```
